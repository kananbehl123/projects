```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)
data['rate']=data['rate'].apply(handleRate)
print(data.head())
```

```python
order_1 = ["Buffet","Cafes","Dining","other"]
sns.barplot(x=data['listed_in(type)'],y=data['rate'],
            order=order_1,
            hue=data['listed_in(type)'],
            palette='icefire',
            saturation = 10)

plt.show()
```

```python
data = pd.DataFrame({
    'rate': ['A', 'B', 'A', 'B', 'C', 'A', 'B', 'C', 'C', 'C']
})

# Create subplots
f, ax = plt.subplots(1, 2, figsize=(7, 3))

# Calculate value counts and plot pie chart
rate_counts = data['rate'].value_counts()
explode = [0.1] * len(rate_counts)  # Create a list with the same length as the number of slices
rate_counts.plot.pie(explode=explode, autopct='%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('Rate')

# Plot countplot
sns.countplot(x='rate', data=data, ax=ax[1])

plt.tight_layout()
plt.show()
```

```python
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': ['X', 'Y', 'Z', 'X', 'Y']  })
numeric_data = data.select_dtypes(include=['number'])

if not numeric_data.empty:
    corr = numeric_data.corr()
    # Plot heatmap
    sns.heatmap(corr, annot=True, cmap='RdYlGn', linewidths=0.2)
    fig = plt.gcf()
    fig.set_size_inches(10, 8)
    plt.show()
else:
    print("No numeric data available to plot.")
```

```python
data = pd.DataFrame({
    'votes': ['Yes', 'No', 'Yes', 'No', 'Yes'],
    'online_order': ['Ordered', 'Not Ordered', 'Ordered', 'Ordered', 'Not Ordered']
})
cross_tab = pd.crosstab(data['votes'], data['online_order'], margins=True)

cross_tab.style.background_gradient(cmap='GnBu')
```

```python
sns.catplot(x='online_order', y='votes', data=data, kind='point')
plt.show()
```
